```python
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline

#Toy data sets
centers_neat = [(-10, 10), (0, -5), (10, 5)]
x_distinct, _ = datasets.make_blobs(n_samples=5000,
                                centers=centers_neat,
                                cluster_std=2,
                                random_state=2)

x_mixed, _ = datasets.make_classification(n_samples=5000,
                                          n_features=10,
                                          n_classes=3,
                                          n_clusters_per_class=1,
                                          class_sep=1.5,
                                          shuffle=False,
                                          random_state=301)
```

```python
plt.style.use('seaborn')
cmap = 'tab10'
```

```python
plt.figure(figsize=(20,8))
plt.subplot(121, title='Distinct Clusters')
plt.scatter(x_distinct[:,0], x_distinct[:,1])
plt.savefig('DistinctCluster.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(20,8))
plt.subplot(122, title='Mixed Cluster')
plt.scatter(x_mixed[:,0], x_mixed[:,1])
plt.savefig('MixedCluster.png', bbox_inches='tight')
```

```python
from sklearn.cluster import KMeans


kmeans_distinct = KMeans(n_clusters=3, random_state=2).fit_predict(x_distinct)
plt.figure(figsize=(20,8))
plt.subplot(121, title='"Distinct" K-Means')
plt.scatter(x_distinct[:,0], x_distinct[:,1], c=kmeans_distinct, cmap=cmap)
plt.savefig('DistinctK-Means.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(20,8))
kmeans_mixed = KMeans(n_clusters=3, random_state=2).fit_predict(x_mixed)
plt.subplot(122, title='"Mixed" K-Means')
plt.scatter(x_mixed[:,0], x_mixed[:,1], c=kmeans_mixed, cmap=cmap)
plt.savefig('MixedK-Means.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(20,8))
kmeans_mixed = KMeans(n_clusters=4, random_state=2).fit_predict(x_mixed)
plt.subplot(122, title='"Mixed" K-Means')
plt.scatter(x_mixed[:,0], x_mixed[:,1], c=kmeans_mixed, cmap=cmap)
plt.savefig('MixedK-Means_4Clusters.png', bbox_inches='tight')
```

```python
# EXAMPLE PLOT
import numpy as np
from scipy.stats import multivariate_normal
import seaborn as sns

x = np.arange(0, 10, 0.1)
y1 = multivariate_normal.pdf(x, mean=5, cov=3)
y2 = multivariate_normal.pdf(x, mean=3, cov=0.8)
sns.set_style("darkgrid")
plt.figure(figsize=(8, 8))
plt.plot(x, y1/2, linewidth=2)
plt.plot(x, y2/2, linewidth=2)
plt.plot(x, (y1+y2)/2, linewidth=2, ls="--")
# plt.ylabel('f(x)', fontsize=14)
# plt.xlabel('x', fontsize=14)
plt.title("Mixture of 1D Gaussians", fontsize=16)
plt.legend(['Gaussian Dist 1', 'Gaussian Dist 2', 'Mixture'])
#plt.legend(['Gaussian Dist 1', 'Gaussian Dist 2' ])
plt.savefig('MixtureGaussian.png', dpi=400)
plt.show()
# print(a.size())
```

```python
from sklearn.mixture import GaussianMixture

gmm_distinct = GaussianMixture(n_components=3).fit(x_distinct).predict(x_distinct)

plt.figure(figsize=(20,8))
plt.subplot(122, title='Distinct GMM')
plt.scatter(x_distinct[:,0], x_distinct[:,1], c=gmm_distinct, cmap=cmap)
plt.savefig('DistinctGMM.png', bbox_inches='tight')
```

```python
from sklearn.mixture import GaussianMixture

gmm_mixed = GaussianMixture(n_components=3).fit(x_mixed).predict(x_mixed)

plt.figure(figsize=(20,8))
plt.subplot(122, title='Mixed GMM')
plt.scatter(x_mixed[:,0], x_mixed[:,1], c=gmm_mixed, cmap=cmap)
plt.savefig('MixedGMM.png', bbox_inches='tight')
```
